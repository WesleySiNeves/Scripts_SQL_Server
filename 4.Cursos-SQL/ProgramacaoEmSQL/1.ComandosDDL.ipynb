# \## O que  collation\_name ? Latin1\_General\_CI\_AS

para saber mais[Clique aqui](https://docs.microsoft.com/pt-br/sql/t-sql/statements/collations?view=sql-server-ver16)

```sql
DECLARE @MeuNome VARCHAR(100) ='Wesley'


if(@MeuNome COLLATE Latin1_General_CS_AS = 'WESLEY')
BEGIN
        SELECT 'Achei'
END
```

```sql
DECLARE @MeuNome VARCHAR(100) ='joo'


if(@MeuNome COLLATE Latin1_General_CI_AS = 'joao')
BEGIN
        SELECT 'Achei'
END
```

## \## Create Database

```sql


-- Drop the database if it already exists
IF  EXISTS (
        SELECT name
        FROM sys.databases
        WHERE name = N'TestNotebookDB'
   )
DROP DATABASE TestNotebookDB
GO

-- Create the database
CREATE DATABASE TestNotebookDB
GO
```
