## <span style="color: rgb(22, 22, 22); font-family: &quot;Segoe UI&quot;, SegoeUI, &quot;Helvetica Neue&quot;, Helvetica, Arial, sans-serif; font-size: 16px; background-color: rgb(255, 255, 255);">O Azure Cosmos DB é um sistema de banco de dados globalmente distribuído que permite que você leia e grave dados de réplicas locais do seu banco de dados e ele os replica de modo transparente para todas as regiões associadas à sua conta do Cosmos.<br><br></span>

## <span style="background-color: rgb(255, 255, 255);">O Azure&nbsp;</span> <span style="background-color: rgb(255, 255, 255); box-sizing: inherit; outline-color: inherit; font-weight: 600;">Cosmos</span> <span style="background-color: rgb(255, 255, 255);">&nbsp;DB é um banco de dados NoSQL totalmente gerenciado e </span> **<mark>foi projetado para fornecer baixa latência, escalabilidade elástica da taxa de transferência, alta disponibilidade e semântica bem definida para consistência de dados.</mark>**

## É possível configurar os bancos de dados para serem distribuídos globalmente e disponíveis em qualquer uma das regiões do Azure. Para diminuir a latência, coloque os dados próximos da localização dos usuários. A escolha das regiões requeridas depende do alcance global de sua aplicação e onde seus usuários estão localizados.

## Bancos de dados do Azure Cosmos DB

# Modo de taxa de transferência provisionada compartilhada<span style="color: rgb(22, 22, 22); font-family: &quot;Segoe UI&quot;, SegoeUI, &quot;Helvetica Neue&quot;, Helvetica, Arial, sans-serif; font-size: 16px; background-color: rgb(255, 255, 255);">: estes contêineres compartilham a taxa de transferência provisionada com outros contêineres no mesmo banco de dados&nbsp;</span>  <span style="color: var(--vscode-foreground);">Contêineres do Azure Cosmos DB</span>

<span style="color: rgb(22, 22, 22); font-family: &quot;Segoe UI&quot;, SegoeUI, &quot;Helvetica Neue&quot;, Helvetica, Arial, sans-serif; font-size: 16px; background-color: rgb(255, 255, 255);">Um contêiner do Azure Cosmos DB é a unidade de escalabilidade para a taxa de transferência provisionada e para o armazenamento. Um contêiner é particionado horizontalmente e, em seguida, é replicado em várias regiões<br></span>  
<span style="color: rgb(22, 22, 22); font-family: &quot;Segoe UI&quot;, SegoeUI, &quot;Helvetica Neue&quot;, Helvetica, Arial, sans-serif; font-size: 16px; background-color: rgb(255, 255, 255);">Ao criar um contêiner, você configura a taxa de transferência em um dos seguintes modos:<br></span>

- # <mark>**Modo de taxa de transferência provisionada dedicada**</mark>: a taxa de transferência provisionada em um contêiner é reservada exclusivamente para esse contêiner e tem o suporte dos SLAs.
    
- # <mark>**Modo de taxa de transferência provisionada compartilhada**</mark>: estes contêineres compartilham a taxa de transferência provisionada com outros contêineres no mesmo banco de dados
