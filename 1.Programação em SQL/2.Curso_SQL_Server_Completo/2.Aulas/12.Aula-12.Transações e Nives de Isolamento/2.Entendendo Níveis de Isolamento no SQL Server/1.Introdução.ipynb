<mark>**Isolamento em bancos de dados.**</mark>

```
O isolamento em bancos de dados é um conceito fundamental relacionado ao controle de concorrência e consistência dos dados em sistemas de gerenciamento de banco de dados (DBMS). Ele se refere à capacidade do sistema de garantir que as operações de leitura e gravação em um banco de dados sejam realizadas de forma segura, mesmo quando várias transações concorrentes estão ocorrendo.

```

  

**ACID (Atomicidade, consistência, isolamento e durabilidade)**

## 

<span style="font-size: 18pt; font-family: Roboto, sans-serif; color: rgb(37, 49, 71); background-color: rgb(255, 255, 0); font-weight: 700; font-variant-numeric: normal; font-variant-east-asian: normal; vertical-align: baseline; white-space: pre-wrap;">ACID é um termo utilizado em ciência da computação para descrever as 4 propriedades</span> <span style="font-size: 12pt; font-family: Arial, sans-serif; color: rgb(51, 51, 51); background-color: rgb(255, 255, 255); font-variant-numeric: normal; font-variant-east-asian: normal; vertical-align: baseline; white-space: pre-wrap;"> que garantem a integridade das transações em um banco de dados. Veja a descrição de cada uma delas:</span>

- <span style="box-sizing: border-box; -webkit-tap-highlight-color: transparent; text-size-adjust: none; -webkit-font-smoothing: antialiased; font-weight: 700; font-size: inherit;">Atomicidade</span>

Essa propriedade garante que todas as transações sejam atômicas (indivisíveis), ou seja, que as transações sejam executadas em sua totalidade. Se ocorrer algum erro, todas as operações que compõem a transação serão descartadas

- <span style="box-sizing: border-box; -webkit-tap-highlight-color: transparent; text-size-adjust: none; -webkit-font-smoothing: antialiased; font-weight: 700; font-size: inherit;">Consistência</span>
    
    ### A execução de uma transação deve levar o banco de dados de um estado consistente para outro estado de consistência, ou seja, toda transação deve respeitar as regras de integridade dos dados (tipo de dado, chave primária etc).
    
    <mark>**O SQL Server fornece uma gama limitada de restrições para impor a integridade lógica, incluindo `PRIMARY KEY`, `FOREIGN KEY`, `CHECK`, `UNIQUE`e `NOT NULL`. Tudo isso tem garantia de satisfação no momento em que uma transação é confirmada. Além disso, o SQL Server garante a integridade _física_ do banco de dados em todos os momentos, é claro.**</mark>

- <span style="box-sizing: border-box; -webkit-tap-highlight-color: transparent; text-size-adjust: none; -webkit-font-smoothing: antialiased; font-weight: 700; font-size: inherit;">Isolamento</span>
    
    É um recurso do banco que tem como objetivo evitar que, em um sistema multiusuário, transações em paralelo interfiram umas nas outras.

- <span style="box-sizing: border-box; -webkit-tap-highlight-color: transparent; text-size-adjust: none; -webkit-font-smoothing: antialiased; font-weight: 700; font-size: inherit;">Durabilidade</span>:
    
    Significa que os efeitos de uma transação são permanentes, podendo ser desfeitos somente como resultado de uma transação posterior e bem-sucedida.
