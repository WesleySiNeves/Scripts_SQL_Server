# <mark>**\# Criação e configuração do Elastc JOB**</mark>

## **<mark>Passo 1) Configurações de Acesso</mark>**

## <mark>1)Executar a criação da master key no banco que hospeda o job</mark>

```sql

		IF NOT EXISTS
		(
			SELECT *
			FROM sys.symmetric_keys
			WHERE name LIKE '%DatabaseMasterKey%'
		)
		BEGIN
			SELECT 'DMK does not exist';
			CREATE MASTER KEY ENCRYPTION BY PASSWORD = 'Minha$enhaSecreta';
		END;
		ELSE
		BEGIN
			SELECT 'DMK exists';
		END;
```

## <mark>**Passo 2)** **Criar a  SCOPED CREDENTIAL**</mark>

##

## <mark>**1) Executar a criação da SCOPED CREDENTIAL  no banco que hospeda o job**</mark>

```sql
SELECT * FROM sys.database_scoped_credentials AS DSC;


DROP  DATABASE SCOPED CREDENTIAL [JobExecuter]

CREATE DATABASE SCOPED CREDENTIAL [JobExecuter] --job_credential 
WITH IDENTITY = 'JobExecuter',
     SECRET = 'Minha$enhaSecreta';
```

## <mark>**2)** **Criar a  LOGIN NO MASTER DB**</mark>

```sql
IF (NOT EXISTS (   SELECT *
                     FROM sys.sql_logins AS SL
                    WHERE SL.name = 'JobExecuter'))
BEGIN
    CREATE LOGIN JobExecuter WITH PASSWORD = 'Minha$enhaSecreta';
END;
```

## <mark>**3)** **Criar a  USER  e associar ao LOGIN NO MASTER DB**</mark>

```sql
/*This script will be executed on master (System database) database */

IF (NOT EXISTS (   SELECT *
                     FROM sys.database_principals AS DP
                    WHERE DP.name = 'JobExecuter'))
BEGIN

    /*Rodar no Mult Script para cada banco configurado*/
    CREATE USER JobExecuter FROM LOGIN JobExecuter;

END;

```

## <mark>**4)** **Criar a  USER  Em todos os bancos que vc precisa executar o JOB**</mark>

```sql
/*Rodar no Mult Script para cada banco configurado*/
IF (NOT EXISTS (   SELECT *
                     FROM sys.database_principals AS DP
                    WHERE DP.name = 'JobExecuter'))
BEGIN

    /*Rodar no Mult Script para cada banco configurado*/
    CREATE USER JobExecuter FROM LOGIN JobExecuter;


END;

ALTER ROLE db_owner ADD MEMBER [JobExecuter];
GO
```

# **<mark>PASSO 3) Criação do target\_groups</mark>**

## <mark>**1) Add a target group  no banco de dados hospedado o JOB**</mark>

```sql
IF(NOT EXISTS (
                  SELECT *
                    FROM jobs.target_groups
                   WHERE
                      target_group_name = 'hml-sql-elastic-job'
              )
  )
    BEGIN
        EXEC jobs.sp_add_target_group 'hml-sql-elastic-job';
    END;

SELECT * FROM jobs.target_groups AS TG;
```

> ## <mark>**2) Add target\_group\_members  no banco de dados hospedado o JOB**</mark>

```sql
IF(NOT EXISTS (
                  SELECT *
                    FROM jobs.target_group_members AS TGM
                   WHERE
                      TGM.server_name = 'rghml-sqlsrv-hm01.database.windows.net'
                      AND TGM.target_type = 'SqlServer'
					  AND TGM.refresh_credential_name ='JobExecuter'
                      
              )
  )
    BEGIN
        EXEC jobs.sp_add_target_group_member 'hml-sql-elastic-job',
                                             @target_type = N'SqlServer',
                                             @refresh_credential_name = N'JobExecuter',
                                             @server_name = 'rghml-sqlsrv-hm01.database.windows.net';
    END;
```

# <mark>**PASSO 4) Criação do JOB**</mark>

## **1) Criar o Job no banco de dados do banco de dados hospedado o JOB**

```sql
IF(NOT EXISTS (
                  SELECT *
                    FROM jobs.jobs AS J2
                   WHERE
                      J2.job_name = 'ManutencaoEPerformace'
                     
              )
  )
    BEGIN

        --Add job for create table
        EXEC jobs.sp_add_job @job_name = 'ManutencaoEPerformace',
                             @description = 'executa as rotinas de manutenção no banco de dados, criação e manutenção de indices e statisticas , alem de expurgos de logs e Elmah';
    END;
```

```sql

SELECT * FROM jobs.jobs AS J
SELECT *
  FROM jobs_internal.jobs AS J
 WHERE
    J.delete_requested_time IS NULL;
```

## **<mark>2) Criar os Steps para o job</mark>**

```sql


IF(NOT EXISTS (
                  SELECT *
                    FROM jobs.jobsteps AS J
                   WHERE
                      J.job_name = 'ManutencaoEPerformace'
                      AND J.step_name = 'uspAutoHealthCheck'
              )
  )
    BEGIN
        EXEC jobs.sp_add_jobstep @job_name = 'ManutencaoEPerformace',
                                 @step_name = 'uspAutoHealthCheck',
                                 @max_parallelism = 5,
                                 @command = N' EXEC HealthCheck.GetSizeDB;',
                                 @credential_name = 'JobExecuter',
								 @retry_attempts  =2,
                                 @target_group_name = 'hml-sql-elastic-job';
    END;
```

## <mark>**\# Detalha as configurações**</mark>

```sql

SELECT J.job_id,
       J.job_version_number,
       J.step_name,
       JD.command_type,
       JD.credential_name,
	   
       JD.target_id,
       JD.initial_retry_interval_ms,
       JD.maximum_retry_interval_ms,
       JD.retry_interval_backoff_multiplier,
       JD.retry_attempts,
       JD.step_timeout_ms,
       JD.max_parallelism,
       CD.text,
	   T.target_group_name
  FROM jobs_internal.jobstep_data AS JD
       JOIN jobs_internal.jobsteps AS J ON J.jobstep_data_id = JD.jobstep_data_id
	   JOIN jobs_internal.targets AS T ON T.target_id = JD.target_id
       JOIN jobs_internal.command_data AS CD ON CD.command_data_id = JD.command_data_id
	   ORDER BY J.job_version_number DESC
```

# **<mark>\# Mostra o Historico de Execuções e Falhas</mark>**

```sql

/* ==================================================================
--Data: 10/02/2021 
--Autor :Wesley Neves
--Observação: Confere a execução o job
 
-- ==================================================================
*/

SELECT j.job_id,
       j.job_name,
       j.job_version,
       j.step_name,
       j.target_type,
       j.step_id,
       j.is_active,
       j.lifecycle,
       j.end_time,
	  current_attempt_start_time = DATEADD(HOUR,-3, j.current_attempt_start_time),
       time_elapsed = DATEDIFF(SECOND, j.start_time, j.end_time),
       tentativas = j.current_attempts,
       j.last_message,
       j.target_resource_group_name,
       j.target_server_name,
       j.target_database_name,
       j.target_elastic_pool_name
  FROM jobs.job_executions j
  WHERE j.lifecycle = 'Failed'
  AND j.step_name ='uspAutoHealthCheck'
  AND j.job_version = (SELECT MAX(JE.job_version_number) FROM jobs_internal.job_executions AS JE)
  AND CAST(DATEADD(HOUR,-3, j.current_attempt_start_time) AS DATETIME2(2)) >= CAST(GETDATE() AS DATE)
```

# **<mark>\# Excluindo banco de dados que não nos interessa</mark>**

```sql

DROP TABLE IF EXISTS #BancosDeDadosExcluidosDoJob;

CREATE TABLE #BancosDeDadosExcluidosDoJob
(
	DatabaseName VARCHAR(250)
)

INSERT INTO #BancosDeDadosExcluidosDoJob (DatabaseName)
VALUES ('auditoria-02.implantadev.net.br_COPY'),
('DNE'),

('Implanta_CRAMA_Homologacao'),
('Implanta_CRABA_Homologacao'),
('Implanta_CFA_Homologacao'),
('Implanta_CONFEA_Homologacao'),
('Implanta.Configuracao'),
('Implanta_CRMVSP_Homologacao'),
('auditoria-27.implantadev.net.br_COPY'),
('cfa-br.implanta.net.br_Antes_Migracao_SISPAT'),
('hml-sqlautomationjobs-db'),
('auditoria-30.implantadev.net.br_COPY')
/* declare variables */
DECLARE @Db_Name_On_Exclude VARCHAR(256);

DECLARE cursor_DB_Name_on_Exclude CURSOR FAST_FORWARD READ_ONLY FOR
SELECT *
  FROM #BancosDeDadosExcluidosDoJob AS BDDEDJ
 WHERE
    BDDEDJ.DatabaseName NOT IN(
                                  SELECT TGM.database_name
                                    FROM jobs.target_group_members AS TGM
                                   WHERE
                                      TGM.membership_type = 'Exclude'
                              );


OPEN cursor_DB_Name_on_Exclude;

FETCH NEXT FROM cursor_DB_Name_on_Exclude
 INTO @Db_Name_On_Exclude;

WHILE @@FETCH_STATUS = 0
    BEGIN
        EXEC jobs.sp_add_target_group_member 'hml-sql-elastic-job',
                                             @membership_type = N'Exclude',
                                             @target_type = N'SqlDatabase',
                                             @server_name = 'rghml-sqlsrv-hm01.database.windows.net',
                                             @database_name = @Db_Name_On_Exclude;

        FETCH NEXT FROM cursor_DB_Name_on_Exclude
         INTO @Db_Name_On_Exclude;
    END;

CLOSE cursor_DB_Name_on_Exclude;
DEALLOCATE cursor_DB_Name_on_Exclude;




SELECT TGM.target_group_name,
       TGM.target_type,
       TGM.membership_type,
       TGM.refresh_credential_name,
       TGM.server_name,
       TGM.database_name,
       TGM.elastic_pool_name
  FROM jobs.target_group_members AS TGM
  WHERE TGM.membership_type ='Exclude'
 ORDER BY
    TGM.membership_type DESC;
```
